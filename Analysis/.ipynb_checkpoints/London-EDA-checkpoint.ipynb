```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import math
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
df = pd.read_csv("London.csv")
print(df.columns)
df
```

```python
df['log_IMD']=np.log(df['IMDScore'])
df['recp_IMD']=np.reciprocal(df['IMDScore'])
df['log_recp_IMD']=np.log(df['recp_IMD'])
df['log_popden']=np.log(df['Pop_Den'])
df['log_workpop']=np.log(df['WorkPop'])
df['Gravity_div_pop']=df['Gravity_Accessibility']/df['TotPop']
df['log_gravity_div_pop']=np.log(df['Gravity_div_pop'])
df['interaction']=df['Log_Gravity']*df['log_popden']
```

```python
sns.histplot(df,x='log_recp_IMD')
```

```python
sns.scatterplot(df, x='Log_Gravity', y='log_IMD',hue='log_popden')
```

```python
sns.residplot(df, x='Log_Gravity', y='log_IMD')
```

```python
X = df[['Log_Gravity','log_popden','interaction']].values
y = df[['log_IMD']].values
model = LinearRegression().fit(X,y)
model.coef_
```

```python
sm_model = sm.OLS(y, X)
res = sm_model.fit()
print(res.summary())
```
